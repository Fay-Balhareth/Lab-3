# **Task 1**
-  Convert a color image (RGB) to (HSV) color space using OpenCV.
- Show the three channels separated.
-  Save them.

```python
# This is necessary for displaying images in Colab.
from google.colab.patches import cv2_imshow
# The cv2 is the main module in OpenCV that provides developers with an easy-to-use interface for working with image and video processing functions.
import cv2

# Read the image, imread() function in cv2 used to read an image from the specified file path.
img = cv2.imread('/content/lol.jpeg')
# Imshow function  used to display an image using OpenCV,            
cv2_imshow(img)
```

```python
# 1. Convert a color image (red, green, blue) to (hue"",saturation "     ", value " ") color space using OpenCV.
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# HSV     
cv2_imshow(hsv_img)
```

```python
# 2. Show the three channels separated.
# 2.1 Display Hue channel,            :=all W :=all H
cv2_imshow(hsv_img[:, :, 0])
# 2.2 Display Saturation channel
cv2_imshow(hsv_img[:, :, 1])
# 2.3 Display Value/Intensity channel
cv2_imshow(hsv_img[:, :, 2])
```

```python
# output above returned to the variables Hue, Saturation, and Value -> so we can use it below without errors.
Hue,Saturation,Value=cv2.split(hsv_img)

# 3. Saving Processed Images In Files.
cv2.imwrite('/content/hue.jpg', Hue)
cv2.imwrite('/content/saturation.jpg', Saturation)
cv2.imwrite('/content/value.jpg', Value)
```

```python
#           
# from google.colab.patches import cv2_imshow
# import cv2
import numpy as np
from matplotlib import pyplot as plt

# img = cv2.imread('/content/lol.jpeg')    

# Get the number of rows and columns of the image, image shape (249, 296, 3 "Channels") -> img.shape[:2] return (249"rows", 296"columns").
num_rows, num_cols = img.shape[:2]
# [no scaling, no rotation, transformation 100 pixels to the right] as X-axis, [no scaling, no rotation, transformation 50 pixels down] as Y-axis,  1 0 100  0 1 50 ->     27  2
translation_matrix = np.float32([[1, 0, 100], [0, 1, 50]])
# Applies a translation transformation to the image.
img_translation = cv2.warpAffine(img, translation_matrix, (num_cols, num_rows))

# Make 2 images side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 5))
ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Before Translation')
ax2.imshow(cv2.cvtColor(img_translation, cv2.COLOR_BGR2RGB))   # RGB     
ax2.set_title('After Translation')
# Display
plt.show()
```
